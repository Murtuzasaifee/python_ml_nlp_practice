```python
# Imports

import pandas as pd
import plotly.express as px
import seaborn as sns
```

Uni Variate Analysis

For single variable

```python
# Load data file 

df = pd.read_csv('/Users/murtuzasaifee/Documents/Personal/Courses/Kirsh Naik/ML and NLP/practice_code/resources/Salary.csv')
```

```python
df.columns
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.info()
```

```python
# Change the data type of column

df['salary'] = df['salary'].astype('float')
df.info()
```

```python
# Cannot change the type of a english word. Python converts the string into binaries

df['city'] = df['city'].astype('str')
df.info()
```

```python
px.histogram(df,x='age')
```

```python
px.histogram(df,x='salary')
```

```python
px.histogram(df, x='experience')
```

```python
# If there is string in Uni Variate Analysis

df['city'].value_counts().plot.bar()
```

Bi Variate Analysis

For two variables

```python
px.bar(df,x='city',y='salary')
```

```python
# Sum of salary of all the employees from the City LA

filtered_df = df[df['city'] == 'LA']
total_la_salary = filtered_df['salary'].sum()
display(total_la_salary)
```

```python
# Average of salary of all the employees from the City LA

df[df['city'] == 'LA']['salary'].mean()
```

```python
px.pie(df,names='city',values='salary')
```

Scatter Plot

It is for finding the relationship

```python
px.scatter(df, x='age',y='salary')

# There is positive correlation between age and salary, but it is less
```

```python
df[['age','salary']].corr()

# There is possitive correlation between age and salary, but it is less which is 0.02
# Correlation can be between only 2 numeric values
```

```python
df[['experience','salary']].corr()

# The correlation between experience and salary is slightly negative, with a correlation coefficient of  -0.0112658
```

```python
sns.heatmap(df[['experience','salary']].corr(),annot=True, cmap='crest')
```

```python
sns.heatmap(df[['age','salary']].corr(),annot=True)
```

```python
df_no_city = df.drop(columns='city')
```

```python
sns.heatmap(df_no_city.corr(), annot=True, cmap='coolwarm')
```

```python
sns.regplot(x='age',y='salary', data=df)
```

```python
sns.regplot(x='experience',y='salary', data=df)
```
