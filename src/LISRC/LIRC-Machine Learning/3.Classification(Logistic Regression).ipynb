```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
df = pd.read_csv("/Users/murtuzasaifee/Documents/Personal/Courses/python_ml_nlp_practice_code/resources/insurance.csv")
```

```python
plt.scatter(df['age'],df['bought_insurance'], color='red',marker='*')
plt.xlabel('Age')
plt.ylabel('Bought Insurance')
plt.title('Age and Bought Insurance')
```

```python
### Features in two dimension
X = df[['age']]

### Target varaible in one dimension
y = df['bought_insurance']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.80,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lrm = LogisticRegression()
lrm.fit(X_train,y_train)
```

```python
y_pred = lrm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test,y_pred)
print(f"Confusion Matric: {conf_matrix}")
```

```python
from sklearn.metrics import classification_report

classi_report = classification_report(y_test, y_pred)
print(classi_report)
```
