```python
# Imports

import pandas as pd
```

```python
# Load data file 

df = pd.read_csv('/Users/murtuzasaifee/Documents/Personal/Courses/python_ml_nlp_practice_code/resources/nba.csv')
```

```python
# Check first 2 rows
df.head(2)

# Check last 2 rows
df.tail(2)

# Check null values count
df.isnull().sum()

# Check the complete info
df.info()

# Check Columns
df.columns

# To get the number of rows and columns
df.shape

```

```python
# Drop the Bonus column
df = df.drop(columns=['Bonus'])

df = df.drop(columns=['Salary'])

df = df.drop(columns='Name')
```

```python
# Drop a row which is completely null

df = df.dropna(how='all', axis=0)

# Drop a column which is completely null

df = df.dropna(how='all', axis=1)
```

```python
# Filling the salary with values

salary_mean = round(df['Salary'].mean(),2)
df['Salary'] = df['Salary'].fillna(salary_mean)
df.head(3)
```

```python
# Filling the Colleage with values

df['College'] = df['College'].fillna('No College')
df.head(5)
```

```python
# Create new data frame with selected columns

new_df = df[['Team', 'College', 'Salary']]
display(new_df)

```

```python
# Get unique values of Team column, get number of unique values of Team Column, get values of all under the Teams column
df['Team'].unique()
df['Team'].nunique()
df['Team'].value_counts()
```

```python
# Same as above

df['College'].unique()
df['College'].nunique()
df['College'].value_counts()
```

```python
# Subtract 2 from the all the age columns

df['Age'].subtract(2)
```

```python
filtered_df = df[df['Team'] == 'Brooklyn Nets']
display(filtered_df)

filtered_df = df[df['College'] == 'Texas']
display(filtered_df)

filtered_df = df[df['Salary'] > 500000]
display(filtered_df)

filtered_df = df[df['Age'] > 30]
display(filtered_df)
```

```python
# Sorting the columns

df.sort_values('Team', ascending=True)

df.sort_values('Team', ascending=False)

df.sort_values('College', ascending=True)

df.sort_values('College', ascending=False)
```

```python
# Top 5 highest salary
df.sort_values('Salary', ascending=False).head(5)

# Age in asencding order
age_df = df.sort_values('Age', ascending=True)
display(age_df)
```
