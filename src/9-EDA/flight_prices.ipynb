```python
import pandas as pd
import sys
import os
import numpy as np
```

```python

# Appending the path for custom modules
sys.path.append(os.path.join(os.getcwd(), '..'))

# Importing Utils from the common_codes module
from common_codes.utils import Utils

df = pd.read_excel(Utils.get_file_path('flight_price.xlsx'))

df.head()
```

```python
df.info()  
```

```python
df.describe()
```

```python
## convert first Date_of_Journery from object to datetime
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])

df.info()
```

```python
## extract day, month and year from Date_of_Journey in separate columns
df['Journey_Day'] = df['Date_of_Journey'].dt.day
df['Journey_Month'] = df['Date_of_Journey'].dt.month
df['Journey_Year'] = df['Date_of_Journey'].dt.year

## convert  into int64
df['Journey_Day'] = df['Journey_Day'].astype('int64')
df['Journey_Month'] = df['Journey_Month'].astype('int64')
df['Journey_Year'] = df['Journey_Year'].astype('int64')

df.head()
```

```python
## Separete Hour and Minutes from Arrival_Time in sperate columns
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute'] = df['Arrival_Time'].str.split(':').str[1]

df['Arrival_Minute'] = df['Arrival_Minute'].apply(lambda x: x.split(' ')[0])

df.drop(['Arrival_Time'], axis=1, inplace=True)

df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)

df.head()
```

```python

df.info()
```

```python
## Separete Hour and Minutes from Dep_Time in sperate columns
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Minute'] = df['Dep_Time'].str.split(':').str[1]

df.drop(['Dep_Time'], axis=1, inplace=True)

df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Minute'] = df['Dep_Minute'].astype(int)

df.head()
```

```python
df.info()
```

```python
df['Total_Stops'].value_counts()
```

```python
df["Total_Stops"].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
## Maximum number of stops are 1

df['Total_Stops'].mode()
```

```python
## Map the nan value with 1 stops as it is the most frequent value
## Also we will change this categoerical value to numerical value with ordinal encoding. As 0 stop will have more prices than 1 stop and so on
## 0 Stop = 5
## 1 Stop = 4
## 2 Stop = 3
## 3 Stop = 2
## 4 Stop = 1
## nan = 4 (1 Stop)

df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 5, '1 stop': 4, '2 stops': 3, '3 stops': 2, '4 stops': 1, np.nan: 4})

df.head()
```

```python
df.info()
```

```python
df.drop(['Route'], axis=1, inplace=True)

df.info()
```

```python
## Separete Hour and Minutes from Duration in sperate columns
# Handle hours
df['Duration_Hour'] = df['Duration'].str.extract(r'(\d+)h').fillna(0).astype(int)

# Handle minutes
df['Duration_Min'] = df['Duration'].str.extract(r'(\d+)m').fillna(0).astype(int)

df['Duration_Hour'] = df['Duration_Hour'].astype(int)
df['Duration_Min'] = df['Duration_Min'].astype(int)

df.drop(['Duration'], axis=1, inplace=True)

df.head()
```

```python
df['Airline'].unique()
```

```python
df['Source'].unique()
```

```python
df['Destination'].unique()
```

```python
## Convert Airline with one hot encoding
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

encoded_values = ohe.fit_transform(df[['Airline','Source','Destination']]).toarray()

encoded_df = pd.DataFrame(encoded_values, columns=ohe.get_feature_names_out(['Airline','Source','Destination']))

df = pd.concat([df, encoded_df], axis=1)

df.drop(['Airline','Source','Destination'], axis=1, inplace=True)

df.head()


```

```python
df.info()
```

```python
df['Additional_Info'].unique()
```

```python
## Convert Additional Info with one hot encoding

# Normalize the column values by converting all strings to lowercase (optional, to handle 'No Info' vs 'No info')
df['Additional_Info'] = df['Additional_Info'].str.lower()

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform the 'Additional_Info' column
encoded_info = encoder.fit_transform(df[['Additional_Info']])

# Convert the result into a DataFrame with appropriate column names
encoded_info_df = pd.DataFrame(encoded_info, columns=encoder.get_feature_names_out(['Additional_Info']))

# Concatenate the original DataFrame with the encoded columns
df = pd.concat([df, encoded_info_df], axis=1)

# Optionally drop the original 'Additional_Info' column
df.drop('Additional_Info', axis=1, inplace=True)
```
