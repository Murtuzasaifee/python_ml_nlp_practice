```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips') ## load the tips dataset from seaborn
tips.head()
```

```python
#Scatter plot

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
# line plot
sns.lineplot(x='size',y ='total_bill', data=tips)
plt.title('Total Bill by Size')
plt.show()
```

```python
## Categorical Plots

sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Total Bill by Day')
plt.show()
```

```python
## Box plot - Helpful to see the outliers

sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
## Violin plot - combination of box plot and KDE plot

sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
## Histogram

sns.histplot(tips['total_bill'], bins=10, kde=True) ## kde is kernel density estimation which makes the curve smooth
```

```python
## KDE plot

sns.kdeplot(tips['total_bill'], fill=True)
```

```python
## pair plot - this will take only numerical columns (feature), will not take categorical columns

sns.pairplot(tips)
```

```python
## Heat Map

corr = tips[['total_bill', 'tip','size']].corr()
sns.heatmap(corr, annot=True)
```
