# SMOTE (Synthetic Minory Oversampling Technique) 


```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_clusters_per_class=1, 
                    weights=[0.90],random_state=12,n_redundant=0)
```

```python
import pandas as pd
import numpy as np

df1= pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
df = pd.concat([df1,df2],axis=1)
df.head()
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['f1'],df['f2'],c=df['target'])
```

```python
from imblearn.over_sampling  import SMOTE
```

```python
## transform the dataset
oversample = SMOTE()
X,y = oversample.fit_resample(df[['f1','f2']],df['target'])
```
