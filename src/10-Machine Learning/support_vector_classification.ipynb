```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
## Create synthetic data points

from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
X
```

```python
y
```

```python
X_df = pd.DataFrame(X)
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')

svc.fit(X_train, y_train)
```

```python
## prediction

y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
## Changing the Syntehtic data to be non-linear

X1,y1 = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
```

```python
sns.scatterplot(x=X1[:,0], y=X1[:,1], hue=y1)
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.25, random_state=42)
```

```python
## Using rfc kernal now


rbf = SVC(kernel='rbf')
rbf.fit(X_train1, y_train1)

y_pred1 = rbf.predict(X_test1)

print(confusion_matrix(y_test1, y_pred1))
print(classification_report(y_test1, y_pred1))
```

```python
## Using polynomial kernal now


polynomial = SVC(kernel='poly')
polynomial.fit(X_train1, y_train1)

y_pred2 = rbf.predict(X_test1)

print(confusion_matrix(y_test1, y_pred2))
print(classification_report(y_test1, y_pred2))
```

```python
## Using Sigmoid kernal now


sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train1, y_train1)

y_pred3 = rbf.predict(X_test1)

print(confusion_matrix(y_test1, y_pred3))
print(classification_report(y_test1, y_pred3))
```

### Hyperparameter tuning with SVC

```python

from sklearn.model_selection import GridSearchCV

## defining parameter range

param_grid = { 'C' : [0.1,1,10,100,1000],
                'gamma' : [1,0.1,0.01,0.001,0.0001],
                'kernel' : ['rbf']}

grid = GridSearchCV(SVC(), param_grid, cv=5, verbose=3, refit=True)
```

```python
grid.fit(X_train1, y_train1)
```

```python
grid.best_params_
```

```python
y_pred4 = grid.predict(X_test1)
print(confusion_matrix(y_test1, y_pred4))
print(classification_report(y_test1, y_pred4))
```
